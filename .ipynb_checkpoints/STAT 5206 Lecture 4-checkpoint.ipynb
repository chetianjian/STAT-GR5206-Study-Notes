# The plot() function.

• plot()函数是R中许多画图函数的基础，时常许多画图函数是分阶段地在plot()的基础上画出

• 每一个plot()函数都新建一个图

• Takes arguments, called graphical parameters, to change various aspects of the plot. (?par)

```r
diamonds <- read.csv("diamonds.csv", as.is = TRUE, fileEncoding="UTF-8-BOM")

diamonds$cut <- factor(diamonds$cut)
diamonds$color <- factor(diamonds$color)
diamonds$clarity <- factor(diamonds$clarity)

set.seed(1)
rows <- dim(diamonds)[1] # rows 为diamonds的行数
diam <- diamonds[sample(1:rows, 1000), ] # sample(1:rows, 1000)随机抽取1000行的行数，然后提取出来存储于diam

rows
head(diam, 5)
```

```r
plot(log(diam$carat), log(diam$price), col = diam$cut)
legend("bottomright", legend = levels(diam$cut), fill = 1:length(levels(diam$cut)), cex = 0.5)
```

```r
colnames(diam)
```

```r
diam
```

```r
plot(log(diam$carat), log(diam$price), col = diam$cut)
legend("bottomright", legend = levels(diam$cut), fill = 1:length(levels(diam$cut)), cex = 0.5)

# 对于每个cut都画出一条regression line
cuts <- levels(diam$cut)
col_counter <- 1

for (i in cuts){
    this_cut <- diam$cut == i
    this_data <- diam[this_cut, ]
    this_lm <- lm(log(this_data$price)~log(this_data$carat))
    abline(this_lm, col = col_counter)
    col_counter <- col_counter + 1}

# 向图中加入一个点 
points(-0.4, 6.8, pch = "*", col = "purple")

# 在刚刚加入的点旁边加上文字
text(-0.4, 6.8 - .2, "New Diamond", cex = .5)
```

# Graph Parameter

pch: 图中点的种类

main: 图的标题

xlab, ylab: 坐标轴的标签

lty: 线的种类，例如"dashed", "dotted"等

lwd: 线的宽度，默认为1

cex: 字体大小，默认为1

xlim, ylim: 坐标轴横跨的范围

mfrow: 将图在一个数组中画出（即并列画出）

col: 画图的颜色

## Change graphical parameters

• Use the par() function to change graphical parameters.

• Change plots per window with mfrow

# ggplot2

• R has several systems for making graphs (we’ve looked at the base R functions).

• ggplot2 is one of the most elegant and flexible.

• ggplot2 uses a coherent system (or ‘grammar’) for describing and building graphs.

We study ggplot2 using the mpg dataset. Let’s try to answer the question:
Do cars with bigger engines use more fuel than cars with small engines?

```r
library('ggplot2')
```

```r
dim(mpg)
head(mpg, 5)
```

```r
ggplot(data = mpg) + geom_point(mapping = aes(x = displ, y = hwy))

# ggplot()创建一个新图， 它同时会创建所需要的坐标系
# 第一个argument是dataset
# 接下来 我们需要向图中加入layers，在这个例子中，geom_point()添加了一个layer of points
# 不同的geom functions具有不同的功能

# geom functions需要mapping argument，它们定义了dataset中变量到可视化特性的映射方式
#这些argument是aes()对的形式，其中x和y指定了哪些变量需要映射到坐标轴上
```

## General structure:
ggplot(data = \<DATA>) + \<GEOM_FUNCTION>(mapping = aes(\<MAPPINGS>))

通过将括号部分替换为一个dataset，一个geom function以及一系列mappings来创建新的图表。

```r
# 如果只输入 ggplot(data = mpg)，会得到什么？
ggplot(data = mpg)
```

```r
# 画hwy vs.cyl的散点图

ggplot(data = mpg) + geom_point(mapping = aes(x = cyl, y = hwy))
```

```r
#  画class vs. drv的散点图，为什么这种图没有什么用处？
ggplot(data = mpg) + geom_point(mapping = aes(x = drv, y = class))
```

# Aesthetic Mappings
• 我们可以向一个散点图中加入第三个变量，将其映射到一个"aesthetic"上

• 一个"aesthetic"是图表中对象的视觉特性，例如尺寸，颜色，点的样式

```r
ggplot(data = mpg) + geom_point(mapping = aes(x=displ, y=hwy, color=class))
```

## Facets
• 可以看到，我们可以通过使用aesthetics来向图表中加入categorical variables

• 还有另一种方法达到同样的效果：将图标分割成facets，facets为子图表(subplot)，每一个subplot展示一个data的subset。

• 使用facet_wrap()来将一个图标根据某一变量进行分割(facet)，所需要输入的argument是公式：
`~variable name`

```r
# nrow控制了输出图表排列的行数
ggplot(data = mpg) +
geom_point(mapping = aes(x = displ, y = hwy)) +
facet_wrap(~ class, nrow = 2)
```

# Geometric Objects

• 我们可以对于各个图表使用各不相同的可视化对象(visual object)来表示数据，这需要通过使用不同的geom来实现

• 一个geom是一个geometrical object，用来在一个图表中表示数据。我们经常通过图表所使用的geom来形容图表本身,例如bar graphs使用bar geoms

```r
# ggplot()不能使用par(mfrow)来规定每行每列多少张图，需要使用gridExtra
library('gridExtra')

plot1 <- ggplot(data = mpg) + geom_point(mapping = aes(x = displ, y = hwy))

plot2 <- ggplot(data = mpg)+ 
geom_smooth(mapping = aes(x = displ, y = hwy))

grid.arrange(plot1, plot2, nrow=1, ncol=2) 
```

• 每一个geom都需要一个mapping argument，但不是任意aesthetic都能应用于某些geom上

• 例如，你可以设置point的shape，但不能设置line的shape，你只能设置line的linetype

• ggplot2拥有大约30种不同的geom

# A few more examples: multiple time series graph
## Time Series Simulation

• The random walk is defined:

$X_{t} = X_{t−1} + \epsilon_{t}, t = 1, . . . , n$
    
• Assume a normal distribution on $\epsilon_{t}$, i.e.,

$\epsilon_{t} \stackrel{\text{i.i.d}}{\sim} N(0, 1), t = 1, . . . , n$

• Plot three random walk realizations.

```r
# Time Series Simulation

# Random Walk
my.randomwalk <- function(n){
  # n <- 100
  X.seq <- rep(NA, n)
  X.seq[1] <- 0
  for (i in 2:n)  {
    X.seq[i] <- X.seq[i-1] + rnorm(1)
  }
  return(X.seq)
}

set.seed(0)
n <- 100
RW.1 <- my.randomwalk(n=n)
RW.2 <- my.randomwalk(n=n)
RW.3 <- my.randomwalk(n=n)
```

```r
RW.new <- data.frame(Time = rep(1:n,3), RW = c(RW.1, RW.2, RW.3), 
                     Sim = c(rep("RW1", n),rep("RW2", n),rep("RW3", n)))

ggplot(data = RW.new) +
geom_line(mapping = aes(x = Time, y = RW, color = Sim))+
labs(title = "Three Random Walks", x = "t", y = "X_t", color="Sim")
```
