Use the built-in iris dataset:

• How many of the iris are species versicolor and have a petal width of less than or equal to 1.2?

• What is the mean petal length of the setosa species iris? (Do this with filtering and with tapply().)

• Make a table of iris species for only those iris with sepal width greater than or equal to 3.0.

• Use the ifelse() command to create a new variable Versicolor that’s an indicator variable (1 if the iris species is versicolor and 0 otherwise). Use the table() function to check your result.

```r
head(iris, 5)
```

```r
# How many of the iris are species versicolor and have a petal width of less than 
# or equal to 1.2?

length(iris[iris$Species == "versicolor" & iris$Petal.Width <= 1.2,])
```

```r
# What is the mean petal length of the setosa species iris? (Do this with filtering 
# and with tapply().)

group <- factor(iris$Species)
Mean_Petal_Length <- tapply(iris$Petal.Length, group, mean)
Mean_Petal_Length[names(Mean_Petal_Length) == "setosa"]
```

```r
# Make a table of iris species for only those iris with sepal width greater than
# or equal to 3.0.

Sepal_Width <- iris[iris$Sepal.Width >= 3.0,]
head(Sepal_Width, 5)
```

```r
# Use the ifelse() command to create a new variable Versicolor that's an indicator
# variable (1 if the iris species is versicolor and 0 otherwise). 
# Use the table() function to check your result.

Versicolor <- ifelse(iris$Species == "versicolor", 1, 0)
head(iris[which(Versicolor == 1),], 10)
```

## Variation

Variation is the tendency of measured values of a variable to change measurement-to-measurement.

• Visualizing the distribution of a variable is the best way to understand the patterns of a variable’s variance.

• Visualize the distribution of a categorical variable using a bargraph.

• Visualize the distribution of a continuous variable using a histogram (or boxplot).

### Bar graphs in R
Produce a bargraph using barplot(height, names.arg) where heights is a vector of values for the heights of each bar and labels is an optional vector of labels for each bar.

• 可以使用table()作为heights的input。

• table()中的顺序与其被输入的factor level的顺序相同。

```r
table(diamonds$cut)
```

```r
names(table(diamonds$cut))
```

```r
barplot(height = table(diamonds$cut), names.arg = names(table(diamonds$cut)))
```

如果我们想按照人数从少到多的顺序排列bar plot怎么办呢？即，按照Fair，Good，Very Good，Premium，Ideal的顺序。

```r
levels(diamonds$cut)
```

```r
diamonds$cut <- factor(diamonds$cut, 
                       level = c('Fair', 'Good', 'Very Good', 'Premium', 'Ideal'))
levels(diamonds$cut)
```

```r
table(diamonds$cut)
names(table(diamonds$cut))
```

```r
barplot(height = table(diamonds$cut), names.arg = levels(diamonds$cut))
```

### Histograms in R

Produce a histogram using hist(values), where values is a vector.

• A histogram divides the x-axis into equally-spaced bins with the height of the bars used to indicate the number of observations falling within the bin.

• Change the width of the histogram's bins using break = argument. This specifies the number of bins.

• Make sure to explore different binwidths as different widths will display different patterns in the data.

```r
# 参数main设置histogram的名称，xlab设置histogram中x轴的名称。

hist(diamonds$carat, main = "Histogram of Carats", xlab = "Carats")
```

```r
# 在这里，breaks = 100代表着分割100次，即生成101个直方。

hist(diamonds$carat[diamonds$carat < 3], 
     breaks = 100, main = "Histogram of Carats", xlab = "Carats")
```

```r
# 也可以随意进行break，在此处，seq(0,3,by=.1)生成从0到3间距为0.1的均匀数据组，在数据
# 点之间生成直方。

hist(diamonds$carat[diamonds$carat < 3],
     breaks = seq(0,3,by=.1), main = "Histogram of Carats", xlab = "Carats")
```

```r
# 此处的分组为：0~0.5, 0.5~0.75, 0.75~1.25, 1.25~2, 2~3

hist(diamonds$carat[diamonds$carat < 3],
     breaks = c(0,.5,.75,1.25,2,3), main = "Histogram of Carats", xlab = "Carats")
```

## Covariation

Covariation is the tendency for the values of two or more variables to vary together in a related way.

• Visualizing the relationship between variables is the best way to spot covariation.

• Visualize the distribution of a categorical variable and a continuous variable using a boxplot.

• Visualize the distribution of two continuous variables using a scatterplot.

## Box plot in R

通过boxplot(values~group)生成boxplot，其中values是需要根据group被分割的vector of data。

• The box stretches from the 25th percentile of the distribution to the 75th percentile (the IQR = Q3 - Q1，即盒子的长度)

• 最中间的线是中位数线。

• The "whiskers" extend to 1.5 times the IQR on either end (Q1 - 1.5IQR and Q3 + 1.5IQR)

```r
boxplot(price~cut, data = diamonds, ylab = "Price", xlab = "Cut")
```

### Scatterplot in R
使用plot(x,y)画散点图，其中x，y分别为代表x值与y值的两个向量。

```r
plot(diamonds$carat, diamonds$price, xlab = "Carats", ylab = "Price ($)")
```

```r
boxplot(carat ~ cut, data = diamonds, ylab = "Carats", xlab = "Cut")
```

# Basics of Plotting

The plot() function.

• The foundation of many of R’s graphics functions.

• Often one builds up the graph in stages with plot()as a base.

• Each call to plot() begins a new graph window.

• Takes arguments, called graphical parameters, to change various aspects of the plot. (?par)

## Building a Visualization: An Example

```r
# First, we create a smaller dataset from diamonds by randomly selecting 1000 rows.

rows <- dim(diamonds)[1]
small_diam <- diamonds[sample(1:rows, 1000), ] #从1到rows（即行数）中随机出1000行
```

```r
plot(log(small_diam$carat), log(small_diam$price), col = small_diam$cut)

# legend:字符或表达式向量，fill:用特定的颜色进行填充，cex:字符大小

legend("bottomright", legend = levels(small_diam$cut),
       fill = 1:length(levels(small_diam$cut)), cex = 0.5)
```

### Adding Lines to a Scatterplot

• Add a straight line with abline(int, slope).

• int is the intercept of the line, and the slope is the slope of the line.

• lines() can also be used.

• Most simply, pass lines() x and y vectors and it connects the points.

```r
plot(log(small_diam$carat), log(small_diam$price), col = small_diam$cut)

# legend:字符或表达式向量，fill:用特定的颜色进行填充，cex:字符大小

legend("bottomright", legend = levels(small_diam$cut),
       fill = 1:length(levels(small_diam$cut)), cex = 0.5)

abline(8, 0, col = "orange", lty = 2)
lm1 <- lm(log(small_diam$price)~log(small_diam$carat))
abline(lm1)
```

Let’s instead plot a regression line for each cut separately.

```r
plot(log(small_diam$carat), log(small_diam$price), col = small_diam$cut)

# legend:字符或表达式向量，fill:用特定的颜色进行填充，cex:字符大小

legend("bottomright", legend = levels(small_diam$cut),
       fill = 1:length(levels(small_diam$cut)), cex = 0.5)

abline(8, 0, col = "orange", lty = 2)
lm1 <- lm(log(small_diam$price)~log(small_diam$carat))
abline(lm1)


cuts <- levels(small_diam$cut)
col_counter <- 1
for (i in cuts) {
    this_cut <- small_diam$cut == i
    this_data <- small_diam[this_cut, ]
    this_lm <- lm(log(this_data$price)~ log(this_data$carat))
    abline(this_lm, col = col_counter)
    col_counter <- col_counter + 1
} 
```

### Adding Points to a Scatterplot

• Most easily done with points(x,y).

• (x,y) is the location of the point to be added.

• example(points) could be helpful.

```r
plot(log(small_diam$carat), log(small_diam$price), col = small_diam$cut)

# legend:字符或表达式向量，fill:用特定的颜色进行填充，cex:字符大小

legend("bottomright", legend = levels(small_diam$cut),
       fill = 1:length(levels(small_diam$cut)), cex = 0.5)

abline(8, 0, col = "orange", lty = 2)
lm1 <- lm(log(small_diam$price)~log(small_diam$carat))
abline(lm1)


cuts <- levels(small_diam$cut)
col_counter <- 1
for (i in cuts) {
    this_cut <- small_diam$cut == i
    this_data <- small_diam[this_cut, ]
    this_lm <- lm(log(this_data$price)~ log(this_data$carat))
    abline(this_lm, col = col_counter)
    col_counter <- col_counter + 1
} 

# 向diamond中加入一个新的点，即$898和0.67 carats

points(-0.4, 6.8, pch = "*", col = "purple")
```

### Adding Text to a Scatterplot

Adding Text to a Scatterplot

• 使用text(x, y, label)向散点图中加入文字，(x, y)是加入文字的位置，label是加入文字的内容。
Most easily done with text(x,y, label).

• The locator() function we saw in Lecture 1 can be useful here.

```r
plot(log(small_diam$carat), log(small_diam$price), col = small_diam$cut)

# legend:字符或表达式向量，fill:用特定的颜色进行填充，cex:字符大小

legend("bottomright", legend = levels(small_diam$cut),
       fill = 1:length(levels(small_diam$cut)), cex = 0.5)

abline(8, 0, col = "orange", lty = 2)
lm1 <- lm(log(small_diam$price)~log(small_diam$carat))
abline(lm1)


cuts <- levels(small_diam$cut)
col_counter <- 1
for (i in cuts) {
    this_cut <- small_diam$cut == i
    this_data <- small_diam[this_cut, ]
    this_lm <- lm(log(this_data$price)~ log(this_data$carat))
    abline(this_lm, col = col_counter)
    col_counter <- col_counter + 1
} 

# 加入一段文字，字体大小为cex = 0.5
text(-0.4, 6.8 - .2, "New Diamond", cex = .5)
```

# Graph Parameter


pch: 图中点的种类

main: 图的标题

xlab, ylab: 坐标轴的标签

lty: 线的种类，例如"dashed", "dotted"等

lwd: 线的宽度，默认为1

cex: 字体大小，默认为1

xlim, ylim: 坐标轴横跨的范围

mfrow: 将图在一个数组中画出（即并列画出）

col: 画图的颜色
